# bar chart in matplotlib

```python
import numpy as np
```

```python
company_name = ['Amazon','Google','Facebook','Apple']
revenue = [1500,1000,900,2000]
profit = [80,300,200,500]
```

```python
ar = np.arange(len(company_name))
ar
```

```python
plt.xlabel('company names')
plt.ylabel('profits in billion')
plt.title('US Tech Giants: Rev and Profit')
plt.xticks(ar,company_name)
plt.bar(ar+0.2,revenue,width = 0.4,label='revenue')
plt.bar(ar-0.2,profit,width=0.4,label='profit')
plt.grid()
plt.legend()
```

```python
plt.ylabel('company names')
plt.xlabel('profits in billion')
plt.title('US Tech Giants: Rev and Profit')
plt.yticks(ar,company_name)
plt.barh(ar+0.2,revenue,label='revenue')
plt.barh(ar-0.2,profit,label='profit')
plt.grid()
plt.legend()
```

## Histogram on Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g')
```

```python

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel('sugar bucket')
plt.ylabel('total patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],color=['green','orange'],label=['men','women']
        )
         #,orientation='horizontal')
plt.legend()
plt.grid()
```

```python
plt.ylabel('sugar bucket')
plt.xlabel('total patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],color=['green','orange'],label=['men','women']
        ,orientation='horizontal')
plt.legend()
plt.grid()
```

# Pie Chart

```python
exp = [15000, 24000, 10000, 15000, 10000]
label = ['Debt','Rent','Food','Clothes','Savings']
```

```python
plt.axis('equal')
plt.pie(exp,labels=label,radius=1.5,autopct='%0.2f%%',shadow=True,explode=[0,0.3,0,0,0])
plt.show()
```
